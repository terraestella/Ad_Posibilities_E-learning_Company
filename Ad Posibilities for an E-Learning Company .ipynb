### Checking for Sample Representativity ###

For the purpose of this analysis, we want to answer questions about a population of new coders that are interested in the subjects the e-learning company teaches. These are, as it has been already said, mostly web and mobile development, but many other domains as well.

We need to find out answers to following questions:
- what subjects interest new programmers most
- how many of them interest in web and mobile development
- what countries are they located in
- what online platforms, free or paid, do they use for learning
- how much money do they have to invest in further online learning
- where do they come from (countries and regions)

```python
# The JobRoleInterest column describes for every participant the role(s) they'd be interested in working in.
# If a participant is interested in working in a certain domain, it means that they're also interested in learning about that domain.
survey_data['JobRoleInterest'].value_counts(normalize = True) * 100
```

It is obvious that many respondents are interested in more than one subject. It would be useful to get a better picture of how many people are interested in a single subject and how many have mixed interests.

In order to do that, we will split each string in the `JobRoleInterest` column to find the number of options for each participant. We'll first drop the null values because we can't split Nan values, and then we'll generate a frequency table for the variable describing the number of options.

```python
non_null_interests = survey_data['JobRoleInterest'].dropna()
split_interests = non_null_interests.str.split(',')

split_interests
```

```python
# to find the number of options for each participant
no_of_options = split_interests.apply(lambda x: len(x))
no_of_options_numb = no_of_options.value_counts()
no_of_options_perc = no_of_options.value_counts(normalize = True) * 100
print(no_of_options_numb)
print()
print(no_of_options_perc)
```

So, the results are:
- 31.65% (2213 participants) have only one subject they are interested in,
- 15.88% (1111 participants) have three subjects they are interested in,
- 15.21% (1064 participants) have four...

This means that vast majority of participants don't have a clear picture what programming niche they want to work in, and/or have mixed interests.

```python
non_null_interests.value_counts()
```

```python
# to find the number of participants who are interested in web/mobile development
web_mob_devs = non_null_interests.str.contains('Web Developer|Mobile Developer')
freq_table = web_mob_devs.value_counts(normalize = True) * 100
print(freq_table)

# frequency table graph
plt.style.use('fivethirtyeight')

freq_table.plot.bar()
plt.title("86% of Participants are Interested\nin Web or Mobile Development", y=1.1)
plt.ylabel("Percentage", fontsize=10)
plt.xticks([0,1], ["Web or Mobile\nDevelopment", "Other Subjects"], rotation=0)
plt.ylim([0,100])
plt.show()
```

86% of the participants are interested in either web or mobile development, which makes this sample representative for our population of interest. The e-learning company wants to advertise their courses to people who are interested in all sorts of programming niches but mostly web and mobile development.

### Participants' Locations and Densities ###

This is where we find out where the participants live and what is their density, that is how many of them are in what country. This is important because ach country can be veiwed as an individual market, so we can easily set our goal to find the two best countries to advertise in.

The column of interest in this part of analysis is `CountryLive` which describes what country each participant lives in (not their origin country). Also, we will include in the analysis only those participants who answered what role(s) they're interested in, so the `JobRoleInterest` column must not have NaN values.

```python
jobRole_nonans = survey_data[survey_data['JobRoleInterest'].notnull()].copy()
abs_freq = jobRole_nonans['CountryLive'].value_counts()
rel_freq = jobRole_nonans['CountryLive'].value_counts(normalize = True) * 100

pd.DataFrame(data = {'Absolute Frequency': abs_freq, 'Percentage': rel_freq})
```

45.7% of our potential customers are located in the US. India, United Kindgdom and Canada follow with 7.7%, 4.6% and 3.8% respectively.

Based on the results, we could say that the two most promissing markets to advertise are United States and India. However, there is one very important piece of information missing, and that is how much money people are actually willing to spend on learning.

### Money for Learning ###

We need to see what amonut of money has each student spent per month since they start learning to code. In order to do that we need to divide `MoneyForLearning` column by `MonthProgramming` column, which describes how many months each sudents spent learning to code.

```python
# MoneyForLearning column - the amount of $ spent from the moment participant started learning
survey_data['MoneyForLearning'].value_counts().head()
```

```python
# replace 0 with 1 to avoid division by 0
jobRole_nonans['MonthsProgramming'].replace(0,1, inplace = True)

# creating new column for amounts of money students spent per month learning
jobRole_nonans['money_per_month'] = jobRole_nonans['MoneyForLearning'] / jobRole_nonans['MonthsProgramming']
# how many null values in the new column
jobRole_nonans['money_per_month'].isnull().sum()
```

```python
# keeping only non-null rows
jobRole_nonans = jobRole_nonans[jobRole_nonans['money_per_month'].notnull()]
jobRole_nonans['money_per_month'].isnull().sum()
```

Removing all the rows that have null values in the `CountryLive` column.

```python
jobRole_nonans['CountryLive'].isnull().sum()
```

```python
jobRole_nonans = jobRole_nonans[jobRole_nonans['CountryLive'].notnull()]
jobRole_nonans['CountryLive'].isnull().sum()
```

```python
jobRole_nonans['CountryLive'].value_counts().head(4)
```

Grouping the remaining data by the CountryLive column and find out how much money a student spends on average each month in the US, India, the United Kingdom and Canada.

```python
mean_by_country = jobRole_nonans.groupby('CountryLive').mean()
mean_by_country['money_per_month'][['United States of America', 'India', 'United Kingdom', 'Canada']]
```

It is suprising, given the socio-economic factors, to see that students in UK and Canada spend less on learning compared to students in India. To eliminate wrong data or outliers coming from wrong survey answers, we should make deeper analysis and visualizations of the problem in had.

### Outliers ###

Visualizing the `money_per_month` column to reveal if there are outliers that mess with our data. 

```python
# countries of interest
four_countries = jobRole_nonans[jobRole_nonans['CountryLive'].str.contains(
    'United States of America|India|United Kingdom|Canada')]

# visualization of distribution
sns.boxplot(y = 'money_per_month', x = 'CountryLive', data = four_countries)
plt.title('Distribution of Money Spent per Month by Country', y = 1.1)
plt.ylabel('Money per Month (USD)')
plt.xlabel('Country')
plt.xticks(range(4), ['US', 'UK', 'India', 'Canada']) # the order of coutries appearing in the column
plt.show()
```

```python
# removing outliers (anything above 20000) and recomputing
jobRole_nonans = jobRole_nonans[jobRole_nonans['money_per_month'] < 20000]
mean_by_country = jobRole_nonans.groupby('CountryLive').mean()
mean_by_country['money_per_month'][['United States of America', 'India', 'United Kingdom', 'Canada']]
```

```python
# countries of interest
four_countries = jobRole_nonans[jobRole_nonans['CountryLive'].str.contains(
    'United States of America|India|United Kingdom|Canada')]

# visualization of distribution
sns.boxplot(y = 'money_per_month', x = 'CountryLive', data = four_countries)
plt.title('Distribution of Money Spent per Month by Country', y = 1.1)
plt.ylabel('Money per Month (USD)')
plt.xlabel('Country')
plt.xticks(range(4), ['US', 'UK', 'India', 'Canada']) # the order of coutries appearing in the column
plt.show()
```

From the graph above we can spot next set of outliers by country:
- US: values over 6000/month
- India: values over 2500/month
- Canada: values over 4000/month

Let's examine these values.

#### US Outliers ####

```python
us_outliers = four_countries[
    (four_countries['CountryLive'] == 'United States of America') &
    (four_countries['money_per_month'] >= 6000)]

us_outliers
```

There are 11 students that paid more than 6000USD for learning to code online. 6 of those attended bootcamps, which can explain the large amount of money spent, but for the remaining 5 we cannot be sure why they spent that kind of money.

The next thing we can see is there are 8 students that spent 3 or less months studying, which could explain the high figures - at the time of the data collected, these students had paid for the whole bootcamp that was going to last for several months.

So, let's remove students that didn't attend bootcamps, as well as those that had been programming 3 or less months at the time of the data collected.

```python
# remove students that didn't attend bootcamps
us_no_bootcamps = four_countries[
    (four_countries['CountryLive'] == 'United States of America') &
    (four_countries['money_per_month'] >= 6000) &
    (four_countries['AttendedBootcamp'] == 0)]
four_countries = four_countries.drop(us_no_bootcamps.index)

# remove students that had been programming 3 or less months
us_study_3_or_less = four_countries[
    (four_countries['CountryLive'] == 'United States of America') &
    (four_countries['money_per_month'] >= 6000) &
    (four_countries['MonthsProgramming'] <= 3)]
four_countries = four_countries.drop(us_study_3_or_less.index)
```

#### India Outliers ####

```python
india_outliers = four_countries[
    (four_countries['CountryLive'] == 'India') &
    (four_countries['money_per_month'] >= 2500)]

india_outliers
```

There is no clue why these students spent that amount of money on learning to code. They might have misunderstood the question or miscalculated their spending figures in some way or the other. Whatever the reason, it seems safer to remove these rows than to leave them and allow them to spoil the whole data.

```python
# removing India outliers
four_countries = four_countries.drop(india_outliers.index)
```

#### Canada Outliers ####

```python
canada_outliers = four_countries[
    (four_countries['CountryLive'] == 'Canada') &
    (four_countries['money_per_month'] >= 4000)]

canada_outliers
```

There's only one participant and, as in US outliers before, he had been programming for no more than two months when he completed the survey. That means he paid a large sum of money in the beginning to enroll in a bootcamp, and probably didn't spend anything aftewards.

```python
# removing Canada outlier
four_countries = four_countries.drop(canada_outliers.index)
```

```python
four_countries
```

```python
# recalculating mean amount of money spent by student each month per country
four_countries.groupby('CountryLive').mean()['money_per_month']
```

```python
# visualization of distribution
sns.boxplot(x = 'CountryLive', y = 'money_per_month', data = four_countries)
plt.title('Money Spent per Month by Country')
plt.xlabel('Country')
plt.ylabel('Money in USD')
plt.xticks(range(4), ['US', 'UK', 'India', 'Canada'])
plt.show()
```
